### van ’t Hoff Equation Derivation

1. **基本关系**  
   对于气相反应，在标准状态下  
   $$
   \Delta_{\mathrm{r}}G_m^{\circ} = -RT \ln K_p
   $$

2. **将吉布斯自由能写成 $H$ 与 $S$**  
   $$
   \Delta_{\mathrm{r}}G_m^{\circ} = 
   \Delta_{\mathrm{r}}H_m^{\circ} - T\,\Delta_{\mathrm{r}}S_m^{\circ}
   $$

3. **把 (2) 代入 (1)，得到**  
   $$
   \ln K_p = -\frac{\Delta_{\mathrm{r}}H_m^{\circ}}{RT} 
             + \frac{\Delta_{\mathrm{r}}S_m^{\circ}}{R}
   $$

4. **对温度求导**  
   （在窄温区内近似认为  
   $\Delta_{\mathrm{r}}H_m^{\circ},\;\Delta_{\mathrm{r}}S_m^{\circ}$ 为常数）  
   $$
   \frac{d\ln K_p}{dT}
   = -\frac{\partial}{\partial T}
     \left(\frac{\Delta_{\mathrm{r}}H_m^{\circ}}{RT}\right) 
   + \underbrace{\frac{1}{R}\frac{d\Delta_{\mathrm{r}}S_m^{\circ}}{dT}}_{\approx 0}
   = \frac{\Delta_{\mathrm{r}}H_m^{\circ}}{R T^{2}}
   $$

   这就是常见的 **van ’t Hoff 方程**。

---

### 📌 关键点回顾
* 步骤 1 用到 $\,\Delta G^{\circ} = -RT\ln K\,$（化学平衡定义）。  
* 步骤 2 用到 $\,\Delta G^{\circ} = \Delta H^{\circ} - T\Delta S^{\circ}\,$（热力学恒等式）。  
* 若需考虑 $\Delta H^{\circ}$ 随 $T$ 变化，可结合 Kirchhoff 方程  
  $$\displaystyle \frac{d\Delta H^{\circ}}{dT} = \Delta C_p^{\circ}\,.$$

---

### 🔄 其他常用形式
* 改写成倒温度导数：  
  $$
  \frac{d\ln K_p}{d(1/T)} = -\frac{\Delta_{\mathrm{r}}H_m^{\circ}}{R}
  $$
  — 绘制 $\ln K_p$ 对 $1/T$ 的图时，斜率即 $-\Delta H^{\circ}/R$。  

* 两温度点积分形式：  
  $$
    \ln\frac{K_{p,2}}{K_{p,1}} 
    = -\frac{\Delta_{\mathrm{r}}H_m^{\circ}}{R}
      \left(\frac{1}{T_2} - \frac{1}{T_1}\right)
  $$
